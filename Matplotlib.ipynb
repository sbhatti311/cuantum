## Simple Line Graph

subplot - more than one in the same figure. fig is the entire figure. ax is the single plot. 

```python
import matplotlib.pyplot as plt
squares = [1,4,5,9,10]
fig,ax=plt.subplots()

#changing label type and line thickness
ax.plot(squares*2, linewidth=3)

ax.set_title("Square", fontsize=12)
ax.set_xlabel('value', fontsize=6)
ax.set_ylabel('square*2', fontsize=8)

ax.tick_params(labelsize=8) # labels on the axes (0,2,4,6)



#other styles

plt.style.use('Solarize_Light2')
plt.show()
```

```python
squares = [1,4,5,9,10]
fig,ax=plt.subplots()
ax.scatter(2,4, s=50) # x,y and size

#changing label type and line thickness
#Formatting the axes
#ax.set_xticklabels(labels)

ax.set_title("Square", fontsize=12)
ax.set_ylabel('square*2', fontsize=8)
ax.set_xlim([0,11]) #x axis range 
ax.set_ylim([-1,11])
ax.set_xticks([2,4,6,8,10]) #x axis ticks
ax.tick_params(axis='both', which='major', labelsize=8)  


plt.plot(range(1, 11), squares*2, '--', color='g', marker=">") 
# "-" (solid line); "-." (dash-dot); ':' (dotted line)
# b (blue) g(green) c(cyan) m(magenta) y(yellow)
# market types are a big list

plt.annotate('first entry', xy=(2, 4)) # for annotating specific point
plt.legend(['First'], loc=4) # legend

plt.show()

plt.savefig('Mysampleplot.png', format='png')
```

```python
x_values = [1,2,3,4]
y_values = [5,3,2,7]

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=20)

ax.axis([0,10,0,20])  ## Sets range for each axis

plt.show()
```

```python
## Colormap
plt.style.use('Solarize_Light2')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Reds, s=10)
plt.show()

# Saving
plt.savefig('square_plot.png', bbox_inches="tight")
```

## histogram

```python
#Histogram

%pip install plotly
import plotly.express as px
import numpy as np
from random import random

poss_results = np.random.randint(1, 11, 20)
poss_results
frequencies = []

for value in poss_results:
    frequency = list(poss_results).count(value)
    frequencies.append(frequency)

print(frequencies)


title="Results"
labels = {'x':'poss_results', 'y':'Frequencies'}
fig = px.bar(x=poss_results,y=frequencies, title=title, labels=labels)
fig.show()
```

## pie chart

```python
import matplotlib_inline

# Flatten the values list to make it 1D
values = np.random.randint(1, 18, 3)
colors = ['b', 'r', 'c']
labels = ["A", 'b', 'c']
explode = (0, 0.2, 0)

plt.pie(values, colors=colors, labels=labels, explode=explode, counterclock=False, shadow=False)
plt.title('Values')

plt.show()
```
